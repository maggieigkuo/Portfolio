#Visualizing Social Media Usage




## Importing Relevant Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Reading the data

```python
social=pd.read_csv("Social Media Usage and Emotional Well-Being.csv")
social
```

## Check for any null values

```python
social.isnull().values.any()
```

## Visualizations

Goal: Produce six visualizations from this data to identify key patterns and trends, effectively conveying a clear narrative about what the data reveals and its implications.

### Visualization 1 -- Average Social Media Usage

This bar graph reveals that Instagram is the most frequently used app in terms of time spent, with users averaging around 150 minutes per day. The second most used app is Snapchat; however, its daily usage is closer in range to Twitter and WhatsApp. Instagram surpasses all other apps by nearly double the average usage time, highlighting its dominance not only among social media platforms but also as a cultural phenomenon. Over the years, Instagram has become a hub for branding, personal image-building, and influencer culture, allowing users to showcase the glamour of their lives. This cultural appeal—offering glimpses into the lives of others, updates from celebrities, and a highly public platform—makes Instagram more engaging and attractive than any other app in the dataset.

```python
social_usage=social.groupby("Platform")[["Daily_Usage_Time (minutes)"]].mean()
sns.barplot(data=social_usage, x="Platform", y="Daily_Usage_Time (minutes)", edgecolor="black", hue="Platform")
plt.xticks(rotation=90)
plt.title("Average Time Usage on Each Social Media Platform")
plt.show()
```

### Visualization 2 -- Average Age and Daily Time Usage on Instagram

After creating the first bar graph, which analyzed the daily time spent on each platform, I became curious about Instagram's popularity and sought to better understand its user demographics. This second bar graph examines the average age of users who spend the most time on Instagram. Interestingly, the graph shows that the age group spending the most time on the app is, on average, 33 years old, followed by 27-year-olds. This finding was surprising, as I initially assumed the average age would skew younger, particularly among teenagers. However, given that the dataset includes users aged 21 to 35, this result aligns with the data. It's plausible that these users are brand owners or content creators, which explains why they dedicate significant time to Instagram. For many of them, it may serve as a full-time job involving marketing, advertisements, and demographic research for their products.

```python
age=social.groupby("Age")[["Daily_Usage_Time (minutes)"]].mean().loc[social.Platform=="Instagram"]
sns.barplot(data=age, x="Age", y="Daily_Usage_Time (minutes)", hue="Age", edgecolor="black", legend=False)
plt.title("Average Age and Daily Time Usage of User's on Instagram")
```

### Visualization 3 -- Correlation Between Time Spent and Dominant Emotion

Another correlation I analyzed was the dominant emotion users experienced while using each platform. Most apps were missing more than one dominant emotion, except for Instagram, which was only missing "boredom". This is likely because Instagram has a significantly larger user base compared to the other apps, providing more data for analysis.

Interestingly, users reported feeling both happiness and anxiety simultaneously when spending extended time on Instagram. This finding can be attributed to the dual pressures of showcasing one's best moments while also comparing oneself to others' posts.

In contrast, LinkedIn had the least amount of reported data. Of the emotions recorded, anxiety and boredom were the most common. This can be explained by the platform's primary purpose: job searching and professional networking. Users without employment may experience anxiety when viewing others' career achievements and advancements, while the app's focused nature can also lead to feelings of boredom.

```python
plt.figure(figsize=(15,10))
sns.barplot(data=social, x="Daily_Usage_Time (minutes)", y="Platform", hue="Dominant_Emotion", edgecolor="black", ci=None)
plt.legend(loc='lower right', fontsize=12)
plt.title("Correlation Between Time Spent on Each Platform and User's Dominant Emotion")
```

### Visualization 4 -- Correlation Between Time Spent and Dominant Emotion

A similar analysis examined gender and their dominant emotions on social media. Stereotypically, young female users are thought to experience the most anxiety and pressure to perform well on these platforms. However, the facet grid reveals that females report feeling the most happiness on social media, challenging this stereotype. This result could be explained by the older age range in the dataset, suggesting that age may influence how women perceive their self-worth and confidence when using social media.

Interestingly, non-binary individuals reported feeling the most neutral on social media. This might be because they experience less societal pressure to conform to traditional gender norms, allowing them to engage with social media without those constraints.

For males, negative emotions such as anxiety, anger, sadness, and boredom were reported at similar levels, but consistently higher than those of non-binary and female users. One possible explanation for men having a less positive experience on social media compared to women could be the lack of male representation on these platforms. Additionally, given the older age range of participants in this study, men in their late 20s and early 30s may not engage with these apps as frequently, potentially contributing to a less fulfilling experience.

```python
color_palette = {
    "Happiness": "red",
    "Sadness": "blue",
    "Neutral": "green",
    "Boredom": "orange",
    "Anxiety": "purple",
    "Anger" : "brown"}
grid=sns.FacetGrid(social,col="Gender", col_wrap=3)
grid.map_dataframe(sns.histplot, x="Dominant_Emotion", binwidth=5, palette=color_palette, hue="Dominant_Emotion")
for ax in grid.axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

### Visualization 5 -- Emotional Correlation Between Likes and Time on Social Media

This scatterplot explores the correlation between users' dominant emotions, daily likes, and time spent on social media. As expected, the more time users spent online and the more likes they received, the happier they reported feeling. Conversely, users who spent less time online and received fewer likes were more likely to experience boredom and sadness. Interestingly, users in the middle range reported higher levels of anger and anxiety compared to other groups.

This pattern suggests that users derive gratification from receiving likes and may feel negative emotions when they fail to meet societal 'standards' for social media engagement. The graph also underscores the compelling nature individuals have to create "on-trend" posts to gain recognition from others, ultimately seeking satisfaction and validation through social media. This highlights the powerful influence of social media and its ability to impact a person's mood.  

```python
plt.figure(figsize=(15,10))
sns.scatterplot(social,x="Daily_Usage_Time (minutes)",y= "Likes_Received_Per_Day", hue="Dominant_Emotion")
plt.legend(loc='upper left', fontsize=12)
plt.plot([0,200],[0,100], '-k')
plt.xlabel("Daily Time Use (Minutes)")
plt.ylabel("Likes Received Per Day")
plt.title("Emotional Correlation Between Likes and Time Usage on Social Media")
plt.grid()
```

### Visualization 6 -- Correlation Between Likes and Posts on Gender

This final visualization demonstrates the prevalence of females on social media. The data shows that the more posts females make in a day, the more likes they tend to receive. In comparison, males and non-binary individuals post less frequently and receive fewer likes. Females are widely distributed across the plot, emphasizing the significant value they place on maintaining their social media presence. This trend also reflects the societal demand for women's presence on these platforms, which conicides with the stereotypical expectation of women "selling" themselves to achieve greater recognition and success.

```python
plt.figure(figsize=(15,10))
color_palette_1 = {
    "Female": "red",
    "Male": "blue",
    "Non-binary": "green"}
sns.scatterplot(social,x="Posts_Per_Day",y= "Likes_Received_Per_Day", palette=color_palette_1, hue="Gender")
plt.legend(loc='upper left', fontsize=15)
plt.plot([0,10],[0,100])
plt.xlabel("Posts_Per_Day")
plt.ylabel("Likes Received Per Day")
plt.title("Correlation Between Likes and Posts on Social Media by Gender")
plt.grid()
```
